```python
file_names = []
data_frames = []
path = "/Users/lunapellegri/Google Drive/Codes-info/digiDAQ-LaBr/LaBr-Traces/2020-11-23/"
for filename in os.listdir(path):
    #print (filename)
    name = os.path.splitext(filename)[0]
    #print (name)
    file_names.append(name)
    df = pd.read_csv(path+filename,sep=' ', header=None)
    #df.columns = ["sampling", "name"]
    df.rename(columns={0: "sampling"}, inplace=True)
    df.rename(columns={1: name}, inplace=True)
    df=df.drop(['sampling'], axis=1)
    data_frames.append(df)

traces = pd.concat(data_frames, axis=1)

#traces.reset_index(inplace=True)
#traces = traces.rename(columns = {'index':'sampling'})
```

```python
traces.head()
```

```python
%matplotlib widget

traces.plot(legend=False)
#plt.scatter(data=traces)
plt.xlabel('Sampling')
plt.ylabel('Amplitude [a.u.]')
plt.title('LaBr Digital DAQ traces')
plt.show()
```
